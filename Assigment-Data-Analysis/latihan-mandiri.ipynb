```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('2019_Yellow_Taxi_Trip_Data.csv')
```

**Data Cleaninng**

```python
meteorites = pd.read_csv('Meteorite_Landings.csv')

condition1 = meteorites['year']>=1960
condition2 = meteorites['year']<=2013

meteorites[condition1 & condition2]
```

```python
met_60_13 = meteorites[condition1 & condition2]
met_60_13.sort_values('year')
```

```python
# untuk setiap tahun, hitung ada berapa meteor
met_count = met_60_13['year'].value_counts().sort_index()
# met_count.sort_values()

met_count.sort_index() # untuk mengurutkan tahun (yang sudah menjadi index)
```

```python
# Visualisasi

met_count.plot()
```

```python
met_count.plot(
    title ='Jumlah Meteor per Tahun',
    xlabel = 'Tahun',
    ylabel = 'Jumlah Meteor'
)
```
