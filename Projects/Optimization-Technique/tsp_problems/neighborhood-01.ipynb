```python
import random

def flipElemen(neighbors):
    selectedIndex = random.randint(0, len(neighbors) - 1)
    if neighbors[selectedIndex] == 0:
        neighbors[selectedIndex] = 1
    else:
        neighbors[selectedIndex] = 0
    return neighbors

def calculateTotal(products, solutions):
    total = 0
    for i in range(len(solutions)):
        if solutions[i] == 1:
            print(products[i][1], products[i][0])
            total += products[i][1]
    return total


products = [
    ["Kacang Garuda", 13800],
    ["Le Minerale", 6000],
    ["Silver Queen", 13000],
    ["Chitato", 14200],
    ["Teh Kotak", 5000],
    ["Susu Ultra", 12000],
    ["Supermie", 3000],
    ["Apel", 12000],
    ["Sari Roti", 5500],
]

solutions = [0, 0, 1, 0, 1, 1, 0, 1, 1]

total = calculateTotal(products, solutions)
print()

flipSolutions = flipElemen(solutions)
print(flipSolutions, "\n")

print("Total Solusi 1:", total)

print("\n=====================================\n")

total2 = calculateTotal(products, flipSolutions)
```

## Neighborhood Diskrit

```python
import random
from itertools import product

def representasiDiskrit(variables):
    ranges = [range(var[0], var[1] + 1) for var in variables]
    return list(product(*ranges))

def oneCompRepresentasiDiskrit(variables):
    ret = []
    for i in range(len(variables)):
        ret.append(random.randint(variables[i][0], variables[i][1]))
    return ret

def getRandomIndexes(variables):
    firstRandomIndex = random.randint(0, len(variables)-1)
    secondRandomIndex = random.randint(0, len(variables)-1)

    while firstRandomIndex == secondRandomIndex:
        firstRandomIndex = random.randint(0, len(variables)-1)
        secondRandomIndex = random.randint(0, len(variables)-1)

    return [firstRandomIndex, secondRandomIndex]

def swappIndex(variables, solusi):
    randomIndexes = getRandomIndexes(variables)
    temp = solusi[randomIndexes[0]]
    solusi[randomIndexes[0]] = solusi[randomIndexes[1]]
    solusi[randomIndexes[1]] = temp
    return solusi

price = 2000000

variables = [
    [1, 5], # PSU
    [1, 8], # MB
    [1, 15], # HDD
    [1, 5], # Monitor
    [1, 3], # Ram
    [1, 10], # Keyboard
]


prices = [
    [50000, 60000, 70000, 80000, 90000], # Prices for PSU
    [100000, 110000, 120000, 130000, 140000], # Prices for WD
    [150000, 160000, 170000, 180000, 190000], # Prices for HDD
    [150000, 160000, 170000, 180000, 190000], # Prices for HDD
]

solusi = representasiDiskrit(variables)

# for s in solusi:
#     total_price = sum(prices[i][s[i] - 1] for i in range(len(s)))
#     print(f"Combination: {s}, Total Price: {total_price}")


print(getRandomIndexes(variables))
neighborhood = swappIndex(variables, solusi)
print("Swapped Solusi:", neighborhood)
```
